```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return 43875 * x + 1500
    
x = np.linspace(0, 10, 100)

y = f1(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, color='blue', label=r'$y = 43875x + 1500$')
plt.title("Gráfica de la función: y = 43875x + 1500")
plt.xlabel("Ventas")
plt.ylabel("UYU$")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def f1(x):
    return 43875 * x + 1500  # Función 1: Ganancias

def f2(x):
    return 42000 * x + 17200  # Función 2: Perdidas

x = np.linspace(0, 20, 200)

y1 = f1(x)
y2 = f2(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, color='blue', label=r'$y = 43875x + 1500$ (Ganancias)')
plt.plot(x, y2, color='green', label=r'$y = 42000x + 17200$ (Pérdidas)')
plt.scatter(8.373, f1(8.373), color='red', label='Punto de equilibrio en x = 8.373')  # Punto de interseccion

plt.title("Ganancias / Pérdidas")
plt.xlabel("Ventas")
plt.ylabel("UYU$")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def f1(x):
    return 43875 * x + 1500

# puntos en x
x1 = 2
x2 = 5

# calcular los puntos correspondientes en la recta
y1 = f1(x1)
y2 = f1(x2)

# calcular la distancia
distance = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

x_vals = np.linspace(x1 - 1, x2 + 1, 100)
y_vals = f1(x_vals)

plt.plot(x_vals, y_vals, label='Segmento Ganancias', color='blue')

plt.scatter([x1, x2], [y1, y2], color='red', zorder=5)
plt.text(x1, y1, f"({x1}, {y1:.2f})", fontsize=10, verticalalignment='bottom')
plt.text(x2, y2, f"({x2}, {y2:.2f})", fontsize=10, verticalalignment='bottom')

plt.plot([x1, x2], [y1, y2], color='green', linestyle='--', label=f'Distancia: {distance:.2f}')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Distancia entre dos puntos del segmento')
plt.legend()

plt.grid(True)
plt.show()

print(f"La distancia entre los puntos ({x1}, {y1}) y ({x2}, {y2}) es: {distance:.2f}")
```

```python
# ecuacion general de las rectas
a1, b1, c1 = 43875, -1, 1500
a2, b2, c2 = 42000, -1, 17200

# formula para el ángulo entre dos rectas
theta = np.degrees(np.arctan(np.abs((a1 * b2 - a2 * b1) / (a1 * a2 + b1 * b2))))

#el metodo .arctan lo que hace es aplicar la inversa a la tan, conocido como tan-1 o arctan

print(f"El ángulo entre las dos rectas es: {theta:.8f} grados")
```

```python
def f1(x):
    return 43875 * x + 1500

def f2(x):
    return 42000 * x + 17200

x = np.linspace(0, 20, 200)

y1 = f1(x)
y2 = f2(x)

plt.plot(x, y1, label='f1(x) = 43875 * x + 1500', color='blue')
plt.plot(x, y2, label='f2(x) = 42000 * x + 17200', color='orange')

# donde y1>y2 hay ganancias y donde y1<y2 hay perdidas
plt.fill_between(x, y1, y2, where=(y1 > y2), color='green', alpha=0.5, label='Region de ganancias')
plt.fill_between(x, y1, y2, where=(y1 < y2), color='red', alpha=0.5, label='Region de perdidas')

plt.xlabel('Ventas')
plt.ylabel('UYU$')
plt.title('Delimitacion de regiones')
plt.legend()
plt.grid(True)

plt.show()
```

```python
def f1(x):
    return 43875 * x + 1500  # ganancias

def f2(x):
    return 42000 * x + 17200  # perdidas

A1, B1, C1 = 43875, -1, 1500  # ecgen
A2, B2, C2 = 42000, -1, 17200  

# Calculamos la distancia desde el origen (0,0) a la recta 1 (ganancias)
distance1 = abs(A1 * 0 + B1 * 0 + C1) / np.sqrt(A1**2 + B1**2)

# Calculamos la distancia desde el origen (0,0) a la recta 2 (perdidas)
distance2 = abs(A2 * 0 + B2 * 0 + C2) / np.sqrt(A2**2 + B2**2)

x = np.linspace(0, 20, 200)

y1 = f1(x)
y2 = f2(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, color='blue', label=r'$y = 43875x + 1500$ (Ganancias)')
plt.plot(x, y2, color='green', label=r'$y = 42000x + 17200$ (Pérdidas)')
plt.plot([0, 0], [0, distance1], color='blue', linestyle='--', label=f'Distancia a Ganancias: {distance1:.2f}')
plt.plot([0, 0], [0, distance2], color='green', linestyle='--', label=f'Distancia a Pérdidas: {distance2:.2f}')

plt.title("Ganancias / Pérdidas")
plt.xlabel("Ventas")
plt.ylabel("UYU$")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
